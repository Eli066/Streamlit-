```python
# Analisis Data Penjualan Sederhana

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Dataset
df = pd.read_csv('database.csv')
df.head()

# 2. Informasi Umum
print(df.info())
print(df.describe())

# 3. Cek nilai kosong
print(df.isnull().sum())

# 4. Visualisasi kolom kategorikal
sns.countplot(x='year', data=df)
plt.title('Distribusi year')
plt.xticks(rotation=45)
plt.show()

# 5. Visualisasi kolom numerik
sns.histplot(df['PM10'], kde=True)
plt.title('Distribusi Jumlah')
plt.show()

# 6. Korelasi
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Matriks Korelasi")
plt.show()

# 7. Simpan versi bersih (jika diperlukan)
df.to_csv('database.csv', index=False)
```
